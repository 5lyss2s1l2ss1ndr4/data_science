### A previsão do clima é gerada através de modelos climáticos, que são sistemas complexos de equações matemáticas que simulam as interações entre a atmosfera, os oceanos, a superfície terrestre e outros componentes do sistema climático. Esses modelos são executados em supercomputadores, que realizam cálculos complexos para simular o comportamento do clima ao longo do tempo.

### O processo de previsão do clima envolve várias etapas:

### Coleta de dados: Dados meteorológicos de várias fontes, como satélites, boias oceânicas, estações meteorológicas terrestres e aviões, são coletados e utilizados para iniciar e alimentar os modelos climáticos.

### Inicialização dos modelos: Os dados coletados são usados para inicializar os modelos climáticos, ou seja, para definir as condições iniciais do sistema climático. Essas condições incluem informações sobre a temperatura, pressão, umidade e outros parâmetros em diferentes partes do mundo.

### Simulação e previsão: Os modelos climáticos então simulam a evolução do sistema climático ao longo do tempo, levando em consideração as interações entre os diferentes componentes, como a atmosfera e os oceanos. Essas simulações geram previsões do clima para diferentes períodos de tempo, que podem variar de alguns dias a várias décadas.

### Pós-processamento e análise: As previsões geradas pelos modelos climáticos são pós-processadas e analisadas para fornecer informações úteis aos meteorologistas e cientistas. Isso inclui a criação de mapas de previsão de temperatura, precipitação, vento, entre outros parâmetros.

### Quanto ao tempo que leva para gerar uma previsão do clima, isso depende da escala temporal da previsão. Previsões de curto prazo (até cerca de duas semanas) são chamadas de previsões meteorológicas e podem ser geradas em questão de horas a partir do momento em que os dados são coletados. Previsões de médio a longo prazo (meses a décadas), conhecidas como previsões climáticas sazonais e decenais, requerem mais tempo para serem geradas devido à complexidade dos modelos e à necessidade de executá-los por períodos mais longos.

### É importante notar que previsões do clima em escalas de tempo mais longas, como a caracterização climática de uma região (como "Tropical úmido" para o sudeste do Brasil), são baseadas em análises históricas de dados climáticos observados ao longo de décadas ou até mesmo séculos. Essas análises identificam padrões e tendências climáticas que podem ser usados para caracterizar o clima típico de uma região específica.

### Para caracterizar o clima de uma região de forma mais precisa e confiável, é comum usar um período de referência de aproximadamente 30 anos. Esse período de 30 anos é conhecido como "período climatológico de 30 anos" e é usado para calcular as médias e variações das condições climáticas ao longo desse intervalo. Esses dados climatológicos de longo prazo permitem identificar os padrões e tendências climáticas típicas de uma região específica.

### A Organização Meteorológica Mundial (OMM), uma agência especializada das Nações Unidas, recomenda o uso de períodos climatológicos de 30 anos para análises climáticas e para definir as médias e extremos climáticos. Isso ajuda a suavizar as variações naturais do clima de curto prazo e a identificar as características climáticas predominantes da região.

### Exemplo: ao afirmar que o sudeste do Brasil possui um clima tropical úmido, essa caracterização é baseada em análises de dados climáticos coletados ao longo de um período de aproximadamente 30 anos ou mais, permitindo uma compreensão mais robusta e confiável das condições climáticas dessa região.

### Para caracterizar o clima de um país, região ou estado, uma série de elementos e condições climáticas são considerados. Esses elementos fornecem informações sobre as características típicas das condições atmosféricas ao longo do tempo. Aqui estão os principais elementos climáticos que são analisados para determinar o clima de uma área:

### Temperatura: A temperatura média e as variações sazonais são fundamentais para caracterizar o clima. Isso inclui temperaturas máximas e mínimas, bem como as médias ao longo de diferentes períodos.

### Precipitação: A quantidade de chuva, neve ou outras formas de precipitação ao longo do ano é crucial para entender os padrões de umidade e secura da região.

### Umidade: A umidade do ar, medida através de parâmetros como a umidade relativa, também é importante. Pode influenciar a sensação térmica e a formação de nuvens e precipitação.

### Pressão atmosférica: A pressão atmosférica varia com as condições climáticas e é fundamental para entender padrões de vento e sistemas de alta e baixa pressão.

### Vento: A velocidade e a direção do vento são elementos significativos para caracterizar o clima, especialmente para áreas costeiras e insulares.

### Radiação solar: A quantidade de energia solar recebida influencia as temperaturas e os padrões climáticos de uma região.

### Nebulosidade: A quantidade de cobertura de nuvens e o tipo de nuvens podem afetar a radiação solar e as temperaturas.

### Altitude: A altitude ou elevação acima do nível do mar também é um fator importante, já que influencia a temperatura e outros elementos climáticos.

### Latitude: A localização em relação à linha do equador influencia a quantidade de energia solar recebida, afetando as temperaturas e as estações do ano.

### Topografia: A forma da superfície terrestre, como montanhas, vales e planícies, pode influenciar os padrões climáticos de uma região, causando variações de temperatura e precipitação.

### Correntes oceânicas: Para regiões costeiras, as correntes oceânicas têm um impacto significativo nas condições climáticas, especialmente nas temperaturas.

### Características geográficas: A presença de grandes massas de água, como oceanos, mares, lagos e rios, pode ter um efeito moderador no clima.

### Variações sazonais: As mudanças climáticas ao longo das estações do ano também são essenciais para caracterizar o clima de uma região.

### Sabendo que todas as colunas, exceto a de data e hora, trabalham com números, iremos fazer os seguintes tratamentos:
#### Substituiremos  (,) por (.);
#### Todas as colunas object serão convertidas para float;
#### A coluna Unnamed e UTC - Hora serão excluídas de todos os arquivos;
#### Converter a coluna Data para datetime;
#### Todos os nulos (NaN) serão substituídos por 0; e
#### Depois salvaremos os arquivos tratados em uma pasta nova.

```python
# Definir a função para aplicar os tratamentos
# Criei a função para chamá-la quando executar o salvamento dos arquivos
def aplicar_tratamentos_converter(precipitacao):
    # Substituir vírgulas por pontos em todas as colunas numéricas
    precipitacao = precipitacao.apply(lambda x: x.replace(',', '.', regex=True) if x.dtype == 'object' else x)

    # Converter colunas object (exceto 'Data' e 'UTC - Hora') para float
    colunas_objeto = precipitacao.select_dtypes(include=['object']).columns
    colunas_objeto_sem_data_e_utc = colunas_objeto.drop(['Data', 'Hora UTC'])
    precipitacao[colunas_objeto_sem_data_e_utc] = precipitacao[colunas_objeto_sem_data_e_utc].apply(lambda x: x.str.replace(',', '.')).astype(float)
    return precipitacao
    
# Chamar a função para cada arquivo
for arquivo in arquivos_csv:
    caminho_arquivo, precipitacao = processar_arquivo(arquivo)
    
    # Aplicar tratamentos ao DataFrame
    precipitacao = aplicar_tratamentos_converter(precipitacao)
    
    # Exibir informações 
    print(precipitacao.info())
    print(precipitacao)
    display(precipitacao)
```

```python
# Definir a função para aplicar os tratamentos
# Criei a função para chamá-la quando executar o salvamento dos arquivos
def aplicar_tratamento_exclusao_colunas(precipitacao):
    colunas_para_excluir = ['Unnamed: 19', 'Hora UTC'] # Não há necessidade mantê-las, meu modelo está baseado na data (DIA)!
    precipitacao = precipitacao.drop(columns=colunas_para_excluir)
    return precipitacao

# Aplicar tratamento ao DataFrame
precipitacao = aplicar_tratamento_exclusao_colunas(precipitacao)

# Exibir informações 
print(precipitacao.info())
print(precipitacao)
display(precipitacao)
```

```python
# Definir a função para aplicar os tratamentos
# Criei a função para chamá-la quando executar o salvamento dos arquivos
def aplicar_tratamento_conversao_data(precipitacao):
    precipitacao['Data'] = pd.to_datetime(precipitacao['Data'])
    return precipitacao

# Aplicar tratamento de conversão da coluna 'Data'
precipitacao = aplicar_tratamento_conversao_data(precipitacao)

# Exibir informações 
print("Após conversão da coluna 'Data':")
print(precipitacao.info())
print(precipitacao)
display(precipitacao)
```

```python
# Substituir valores nulos por 0
# Lembrando que estes dados que estão NaN (nulos), são dados que naquele determinado dia não aconteceu nenhum evento climático
# Exemplo: Dia 20JULHO2023, não choveu o dia inteiro!!!
def aplicar_tratamento_substituicao_nulos(precipitacao):
    precipitacao = precipitacao.fillna(0)
    return precipitacao

# Aplicar tratamento de substituição de valores nulos por 0
precipitacao = aplicar_tratamento_substituicao_nulos(precipitacao)

# Exibir informações
print(precipitacao.info())
print(precipitacao)
display(precipitacao)
```

Por fim, tira a média desses erros quadrados para dar o erro médio quadrático.

```python
kf = KFold(n_splits=8) 

mse_per_fold = []
dfs = []

for train_index, test_index in kf.split(X):
    
    # Separando os dados de treino e teste para cada fold específico
    X_train, X_fold = X[train_index], X[test_index]
    y_train, y_fold_target = y[train_index], y[test_index]

    # Treinando o modelo para cada fold
    modelo_gb.fit(X_train, y_train)

    # Fazendo as previsões para cada fold
    y_fold_pred = modelo_gb.predict(X_fold)

    # Calculando o MSE para cada fold
    mse = mean_squared_error(y_fold_target, y_fold_pred)
    mse_per_fold.append(mse)

    # Criando um dataframe para analisar os resultados de cada fold
    df_fold = pd.DataFrame({'target': y_fold_target, 'pred': y_fold_pred})
    df_fold['erro'] = df_fold['target'] - df_fold['pred']
    dfs.append(df_fold)

# Assumindo que dfs é sua lista de DataFrames para cada fold
for i, df in enumerate(dfs, 1):
    print(f"Fold {i}:")
    display(df.sort_values(by='erro'))
```

```python
# Verificando os outliers em cada coluna (X)
colunas = [
    "TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)",
    "UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)",
    "VENTO, VELOCIDADE HORARIA (m/s)",
    "TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)",
    "TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)",
    "VENTO, DIREÇÃO HORARIA (gr) (° (gr))",
    "VENTO, RAJADA MAXIMA (m/s)",
    "UMIDADE RELATIVA DO AR, HORARIA (%)",
    "TEMPERATURA DO PONTO DE ORVALHO (°C)",
    "UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)",
    "TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)",
    "RADIACAO GLOBAL (Kj/m²)",
    "TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)",
    "PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)",
    "PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)",
    "PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)"
]

for col in colunas:
    # Usando desvio padrão para detectar outliers
    mean = np.mean(precipitacao[col])
    std = np.std(precipitacao[col])
    outliers = precipitacao[(precipitacao[col] < mean - 2 * std) | (precipitacao[col] > mean + 2 * std)]
    print(f"Outliers para {col} usando desvio padrão:", outliers)

    # Usando IQR para detectar outliers
    Q1 = np.percentile(precipitacao[col], 25)
    Q3 = np.percentile(precipitacao[col], 75)
    IQR = Q3 - Q1
    outliers = precipitacao[(precipitacao[col] < Q1 - 1.5 * IQR) | (precipitacao[col] > Q3 + 1.5 * IQR)]
    print(f"Outliers para {col} usando IQR:", outliers)
    
# Não irei excluir os outliers, pois necessitamos de todas as colunas, independentemente do laço temporal.
```
